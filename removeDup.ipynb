```python
import pandas as pd

# List to store all DataFrames
all_dataframes = []

# Loop through each file
for i in range(1, 3):
    #Change this accordingly
    file_name = f'Sheet{i}_go.xlsx'
    
    # Read the Excel file into a DataFrame
    df = pd.read_excel(file_name)
    df = df[df['Title'] != "Title not available"]
    
    # Append the DataFrame to the list
    all_dataframes.append(df)

# Concatenate all DataFrames into one DataFrame
combined_df = pd.concat(all_dataframes, ignore_index=True)

# Find duplicates within the 'Title' column of the combined DataFrame
duplicates = combined_df[combined_df.duplicated(subset=['Title'], keep=False)]
# duplicates = duplicates[duplicates.duplicated(subset=['Title'], keep=False)]
duplicates = duplicates.drop_duplicates(subset=['Title'], keep='first')

# Loop through each DataFrame and remove duplicates based on the 'Title' column
for i, df in enumerate(all_dataframes):
    all_dataframes[i] = df[~df['Title'].isin(duplicates['Title'])]

dup_file_name = f'duplicates.xlsx'
duplicates.to_excel(dup_file_name, index=False)
# # # Save each DataFrame to a new Excel file without duplicates
for i, df in enumerate(all_dataframes):
    #Change this accordingly
    new_file_name = f'Sheet{i+1}_go_without_duplicates.xlsx'
    df.to_excel(new_file_name, index=False)
    
```
